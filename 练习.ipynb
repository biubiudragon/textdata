**误差棒数据偏离情况**

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
fig=plt.figure(figsize=(8,6))
plt.boxplot(data,sym='s',vert=True)
plt.xticks(ticks=[y+1 for y in range(len(data))],labels=['x1','x2','x3'])##ticks每个点的起始位置
plt.xlabel('x')
plt.title('box plot')
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
data=[np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')
axes[1].boxplot(data)
axes[1].set_title('box plot')
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(data))],['x1','x2','x3','x4'])
plt.show()
```

**三维图**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
##3D图
ax=fig.add_subplot(121,projection='3d')
plt.show()
```

```python
##设置随机种子
np.random.seed(1)
def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.randn(n)+vmin
fig=plt.figure()
##开始绘制
ax=fig.add_subplot(111,projection='3d')
n=100

##颜色和标记以及取值范围
for c,m,zlow,zhigh in [('r','o',-50,-25),('b','x','-30','-5')]:
    xs=randrange(n,23,32)
    ys=randrange(n,0,100)
    zs=randrange(n,int(zlow),int(zhigh))
    ax.scatter(xs,ys,zs,c=c,marker=m)
    ax.view_init(45)##不同视角
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for c,z in zip(['r','g','b','y'],[30,20,10,0]):
    xs=np.arange(20)
    ys=np.random.rand(20)
    cs=[c]*len(xs)
    ax.bar(xs,ys,zs=z,zdir='y',color=cs,alpha=0.5)
plt.show()
```
